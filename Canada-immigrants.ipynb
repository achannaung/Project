```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/achannaung/DV-Final_Assignment/main/canada-immigration.csv')
df.head()
```

```python
dfc = df.groupby('Continent', axis=0).sum()
dfc
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colorlists = ['skyblue','gold','crimson', 'lightgreen', 'yellow', 'lightcoral']
explodelist = [0.1, 0, 0 ,0.1,0.2, 0.5]

dfc['Total'].plot(kind='pie',
                 figsize=(5, 9),
                 autopct= '%1.1f%%',
                 startangle=50,
                 shadow=True,
                  label=None,
                  pctdistance=0.75,
                colors= colorlists,
                  explode = explodelist
                 )
plt.title('Number of Immigrant to Canada by Continents from 1980 to 2013')
plt.axis('equal')

plt.legend(labels= dfc.index, loc='lower left')
plt.show()

```

```python
df.set_index('Country', inplace=True)
df.head()
```

```python
mm = df.loc[['Myanmar']]
mm.drop(['Continent', 'Region'], axis=1, inplace=True)
mm
```

```python
mm = mm.transpose()
mm
```

```python
mm.plot(kind='line',
       figsize=(6,4),
       color='red',
       )

plt.title('Total Immigrants from Myanmar to Canada [1980-2013]')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.legend(labels=mm)
plt.show()
```

```python
asia = df[df['Continent'] == 'Asia']
ast = asia.groupby('Country', axis=0).sum()
ast.head()
```

```python
asean = df[df['Region'] == 'South-Eastern Asia']
asean = asean.groupby('Country', axis=0).sum()
asean
```

```python
colorlists = ['tomato', 'lime', 'lightblue', 'orangered', 'gold', 'crimson','Royalblue', 'salmon', 'magenta', 'red' ]
explodelists = [0,0,0.2,0.2,0.1,0.2,0,0,0,0]

asean['Total'].plot(kind='pie', 
                   figsize=(5,8),
                    autopct='%1.1f%%',
                    pctdistance=0.5,
                    shadow=True,
                    startangle=45,
                    labels=None,
                    colors=colorlists,
                    explode= explodelists,
                   )
plt.legend(labels=asean.index, loc='lower right')
plt.show()
```

```python
asean.sort_values('Total', ascending=False)
```

```python
phmm = asean.loc[['Philippines','Viet Nam', 'Malaysia', 'Singapore', 'Indonesia'], years].transpose().sum()

explodel = [0.1,0,0,0.1,0.15]
phmm.plot(kind='pie',
         figsize=(5, 8),
         startangle=30,
         shadow=True,
         autopct='%1.1f%%',
         labels=None,
          explode= explodel,
         )

plt.legend()
plt.show()
```

```python
hg.rename(columns={'stays_in_weekend_nights': 'weekend', 'stays_in_week_nights':'week'}, inplace=True)
hg
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

hg['weekend'].plot(kind='bar', figsize=(10,3), ax=ax0, color='deeppink')
hg['week'].plot(kind='line',figsize=(10,3),ax=ax1, color='red')

plt.show()
```
